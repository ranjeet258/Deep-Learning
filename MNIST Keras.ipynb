```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
keras.datasets.mnist.load_data()
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))  # 1D convert ker dega
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))

model.add(Dense(10,activation='softmax'))  # 1 se jyada class in clasification ...softmox
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history= model.fit(X_train,y_train, epochs= 25, validation_split=.2 )
```

```python
y_prob= model.predict(X_test)
```

```python
y_prob
```

```python
y_pred= y_prob.argmax(axis=1)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylim(0, 0.3) # You can adjust these values as needed
plt.xlim(0, 35) # You can adjust these values as needed
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
